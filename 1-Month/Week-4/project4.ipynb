```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Age': [22, 25, 47, 52, 46, 56, 55, 60, 32, 40],
    'Salary': [25000, 32000, 70000, 80000, 62000, 90000, 85000, 95000, 40000, 52000],
    'Experience': [1, 2, 20, 25, 23, 30, 28, 35, 5, 15],
    'Department': ['HR', 'HR', 'Tech', 'Tech', 'Tech', 'Management', 'Management', 'Management', 'HR', 'Tech']
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Dataset:\n", df.head())
```

```python
sns.set(style="whitegrid")
```

```python
# Scatter Plot: Age vs Salary
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Age', y='Salary', hue='Department', data=df)
plt.title("Scatter Plot: Age vs Salary")
plt.show()
```

```python
# Histogram: Salary distribution
plt.figure(figsize=(6, 4))
sns.histplot(df['Salary'], bins=5, kde=True)
plt.title("Histogram: Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot: Salary by Department
plt.figure(figsize=(6, 4))
sns.boxplot(x='Department', y='Salary', data=df)
plt.title("Boxplot: Salary by Department")
plt.show()
```

```python
#  Heatmap: Correlation between numeric features
plt.figure(figsize=(6, 4))
correlation = df[['Age', 'Salary', 'Experience']].corr() 
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Heatmap: Feature Correlations")
plt.show()
```
